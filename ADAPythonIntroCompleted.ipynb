## Loading data
- Data is a current clamp recording from a mitral cell in the olfactory bulb, a current injection is given to evoke action potential firing
- We are going to plot the raw data, use event detection to automatically identify spikes and then calculate the average firing rate of the cell
- See np.genfromtxt for how to load different formats or sections of files https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html
- If you are dealing with excel type data the pandas library is very good for loading and displaying excel files

```python
data = np.genfromtxt('mitralSpikes.txt');
dataPoints = len(data)
peak = np.max(data)
AHP = np.min(data)

print("this data has ",dataPoints,'data points, with a peak of ',peak,'and a minimum of ',AHP)
```

**Now we will visualise it**

```python
plt.plot(data); # matplotlib enables very quick and simple display of data
```

**To make the plots more accurate we need an xscale and to add axis labels**

```python
sampleRate = 15000
xScale = np.arange(0,(len(data)/sampleRate),1/sampleRate)
```

```python
fig, ax = plt.subplots(1,figsize=(15,5)) 
ax.plot(xScale,data)

ax.set(xlabel='Time (s)',ylabel='Vm (mV)');

```

**now lets get the time of each action potential so that we can calculate spike rate**
- The Scipy library provides lots of fancy functionality
- We will use it to find the peaks i.e. spikes in this signal

```python
import scipy.signal as signal
```

```python
peaks, props = signal.find_peaks(data,height=0) #this function returns two things: 1) peaks
# which is the location of the spike in sample points and props which contains other details about the peaks
peaks=peaks/sampleRate  # convert the sample points to time

print('we detected ',len(peaks),' spikes')
```

**We have a list of spike times, we can now make a histogram of the instantaneous spike rates**

```python
interSpikeIntervals = peaks[1:]-peaks[0:-1]  # subtracts 

meanISI = np.mean(interSpikeIntervals)

fig, ax = plt.subplots(1)
ax.hist(1/interSpikeIntervals);
ax.set(xlabel='Hz',ylabel='Events')
print('The mean spike rate is',round(1/meanISI,2),'Hz')
```

```python
peakY=np.zeros(len(peaks))     
fig, axs = plt.subplots(1, 2, figsize=(12, 5),constrained_layout=True) # create a new canvas 
#for the figure with a constrained layout and two subplots

axs[0].plot(xScale,data,label='raw data');
axs[0].plot(peaks,peakY,'.',label='detected spikes');
axs[0].set(xlabel='Time (s)',ylabel='Vm (mV)',ylim=(-70,35));

axs[0].legend(loc='upper right')

axs[1].hist(1/interSpikeIntervals);
axs[1].set(xlabel='Hz',ylabel='Events', title='Instantaneous spike rate');

```

## A more complex figure 

- This data is a voltage clamp recording of a Kv2 potassium current taken from: [Johnston, et al (2008)](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2538803/)

- We are going to plot the raw data with the command voltage and make plots of the current voltage relationship and the activation kinetics of the channel

```python
#Load the data
measuredI=np.loadtxt('VClampI.txt', delimiter=','); ## load some new data, this is comma seperated...
commandV=np.loadtxt('VClampV.txt', delimiter=','); ## load some new data, this is comma seperated...
VCsampleRate = 10000
xScaleVC = np.arange(0,len(measuredI)/VCsampleRate, 1/VCsampleRate)
# the units of measuredI is pA
# the units of commandV is mV
```

```python
# check what its dimensions look like
print('len returns',len(measuredI))
print('whereas .shape returns the length of each dimension')
print(measuredI.shape) # standard python index for multiple dimension is [z,x,y]
measuredI.shape[1]
```

```python
plt.plot(measuredI);
```

**All of the current traces are in pA lets convert them all to nA by dividing them all by 1000**

```python
measuredI=measuredI/1000
plt.plot(measuredI);
```

**Building a more complex figure**

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharex='col',constrained_layout=True)

axs[0,0].plot(xScaleVC,measuredI)
axs[0,0].set(ylabel='current (nA)',xlim=(.72,.9))
# add lines to indicate where measurements were taken for I/V relationship and activation
axs[0,0].axvline(.8,0.1,1,linestyle='--')
axs[0,0].axvline(.8724,0.1,1,linestyle='--',color='r')

axs[1,0].plot(xScaleVC,commandV)
axs[1,0].set(ylabel='Voltage (mV)', xlabel='Time (ms)')


# below we are plotting the value at point 7800 (where the dashed vertical line is) from each colum of the 
# current data against the voltage at the same point to generate the current voltage relationship
axs[0,1].plot(commandV[8000,:],measuredI[8000,:],'--D',linewidth=2) # note the short hand for specing 
#the style of the line and marker
axs[0,1].set(title='Current voltage relationship',ylabel='current (nA)')

## below we are calculating the voltage dependent activation by measuring the tail current 
#(dashed vertical red line) similar to above. We then normalise it by subtracting the last value [-1]
# corresponding to the most negative voltage and then dividing by the current at the peak voltage index 0
activation = measuredI[8724,:]-measuredI[8724,-1]
activation=activation/activation[0]

axs[1,1].plot(commandV[8000,:],activation,'r-o',linewidth=2)
axs[1,1].set(title='Activation curve',ylabel='I/Imax',xlabel='Voltage (mV)');
```

## Defining and fitting functions
**An experimenter is measuring the amonia on the breath of a patient having kidney dialysis and wants to determine the time constant of clearance of urea/amonia from the body**
- Samples were taken every minute
- each data point is the concentration of amonia in mg/dl 

```python
import scipy.optimize as fit  ## imports the optimization tool box as fit
breath = np.load('decay.npy')
```

> 1) Plot the data as a scatter plot i.e. points for markers rather than lines between points.  
> 2) Define the appropriate equation to describe clearance from the body.  
> 3) Fit this equation to your data.    
> 4) Determine the time constant of urea clearance and print this with some descriptive text


```python
xScale =np.arange(0,breath.shape[0],1)
plt.plot(xScale,breath,'o', label='raw data');
plt.xlabel('Time (mins)')
plt.ylabel('Ammonia on Breath (mg/dl)');

def decay(x,a,tau): # the 1st input to your function should be the dependent variable 
    return a*np.exp(x/-tau)  # this is the equation for an exponential decay

constants, coeffs = fit.curve_fit(decay,xScale,breath)  # this uses scipy's optimize module to fit the equation to the data
plt.plot(xScale,decay(xScale,constants[0],constants[1]), label='fit')
plt.legend()
print('The decay of urea during dialysis had a time constant of',constants[1],'minutes')
```
